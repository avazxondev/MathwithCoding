```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the function f(x, y) = (y + sqrt(x*y)) / x
def f(x, y):
    return (y + np.sqrt(x * y)) / x
```

```python
# Euler's method
def euler_method(f, x0, y0, h, n):
    x_values = [x0]
    y_values = [y0]

    for i in range(n):
        y0 = y0 + h * f(x0, y0)
        x0 = x0 + h
        x_values.append(x0)
        y_values.append(y0)

    return x_values, y_values
```

```python
# 4th-order Runge-Kutta method
def runge_kutta_method(f, x0, y0, h, n):
    x_values = [x0]
    y_values = [y0]

    for i in range(n):
        k1 = h * f(x0, y0)
        k2 = h * f(x0 + h/2, y0 + k1/2)
        k3 = h * f(x0 + h/2, y0 + k2/2)
        k4 = h * f(x0 + h, y0 + k3)

        y0 = y0 + (k1 + 2*k2 + 2*k3 + k4) / 6
        x0 = x0 + h
        x_values.append(x0)
        y_values.append(y0)

    return x_values, y_values
```

```python
# Initial conditions
x0 = 1  # Starting x-value
y0 = 1  # Starting y-value
h = 0.1  # Step size
n = 50   # Number of steps

# Compute solutions using Euler and Runge-Kutta methods
x_euler, y_euler = euler_method(f, x0, y0, h, n)
x_runge_kutta, y_runge_kutta = runge_kutta_method(f, x0, y0, h, n)
```

```python
# Plotting the results
plt.plot(x_euler, y_euler, label="Euler Method", marker='o')
plt.plot(x_runge_kutta, y_runge_kutta, label="Runge-Kutta Method", marker='x')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Approximate Solution of the Differential Equation")
plt.legend()
plt.show()
```
