```python
import numpy as np
import matplotlib.pyplot as plt

# Initial conditions and parameters
x0 = 0        # Initial x value
y0 = 1        # Initial y value
x_end = 5     # Final x value
h = 0.1       # Step size

# Create arrays for x and y values
x_vals = np.arange(x0, x_end + h, h)
y_vals = np.zeros(len(x_vals))
y_vals[0] = y0

# Implement the finite difference scheme
for i in range(1, len(x_vals)):
    x_mid = (x_vals[i] + x_vals[i-1]) / 2    # Midpoint of the interval
    f_mid = x_mid                            # f(x) = x, so f_mid = x_mid
    y_vals[i] = y_vals[i-1] + f_mid * h      # Update y using the scheme

# Plot the results
plt.plot(x_vals, y_vals, label="Finite Difference Solution")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Solution of Differential Equation using Finite Difference Scheme")
plt.legend()
plt.grid()
plt.show()
```
